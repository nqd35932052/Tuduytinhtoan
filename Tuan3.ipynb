Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình

```python
a,b,c = map(int,input().split("/"))
if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
    print("Không tồn tại tam giác")  
else:
    #Phân loại tam giác
    if a == b == c :
        loai = "Tam giác đều"
    elif a == b or a == c  or b == c:
        if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        loai = "Tam giác vuông" 
    else:
        loai = "Tam giác thường" 
    P = a + b + c
    p = P/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5

    print(f"Loại tam giác: {loai}")
    print(f"Chu vi = {p}")
    print(f"Dien tich = {s}")

```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax^2 + bx^2 + c = 0)


```python
a,b,c=map(int,input("Enter 2 numbers").split("/"))
if a == 0:
    if b == 0:
        if c == 0:
            print("Infinitive solutions")
        else:
            print("No solutions")
    else:
        x = -c/b
        print(f"x = {x}")
else:
    delta = b**2-4*a*c
    if delta < 0:
        print("No solutions")
    elif delta == 0:
        print(f"x = {-b/(2*a)}")
    elif delta > 0:
        x1 = (-b + delta**0.5)/2*a
        x2 = (-b - delta**0.5)/2*a
        print(f"x1 = {x1} - x2 = {x2}")

```

Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bànphím).

Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
a,b,m,c,d,n=map(float,input("Nhập các giá trị: ").split("/"))
Det=a*d-b*c 
Detx = m*d-n*b
Dety = a*n-c*m
if Det == 0:
    if Detx ==  Dety == 0:
        print("Infinitive Solution")
    else: 
        print("No solution")
else:
    x = Detx/Det
    y = Dety/Det
    print(f"x = {x}")
    print(f"y = {y}")

```
