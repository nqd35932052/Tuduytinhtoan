Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường". 


```python
a,b,c = map(int,input().split("/"))
if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
    print("Không tồn tại tam giác")  
else:
    #Phân loại tam giác
    if a == b == c :
        loai = "Tam giác đều"
    elif a == b or a == c  or b == c:
        if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        loai = "Tam giác vuông" 
    else:
        loai = "Tam giác thường" 
    P = a + b + c
    p = P/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5

    print(f"Loại tam giác: {loai}")
    print(f"Chu vi = {p}")
    print(f"Dien tich = {s}")
```

Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 

```python
n = int(input())
bac = [50,50,]
gia = [1500,2000,3000]
tong = 0
for i in range(len(bac)):
    x = min(n, bac[i])
    tong += x*gia[i]
    n -= x
if n > 0 :
    tong += n*gia[-1]

print(tong)

```
